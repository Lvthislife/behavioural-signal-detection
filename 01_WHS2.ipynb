# 🔍 WHS NLP Capstone: Predicting Effort and Action Quality
This notebook performs:
- Data load and validation
- Exploratory Data Analysis (EDA)
- TF-IDF and RoBERTa-based classification for `effort` and `action_quality`
- Explainability using SHAP
- Persona-wise comparison and trend diagnostics
- Ensures proper feature extraction before train/test split

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# Load dataset
df = pd.read_csv('synthetic_whs_incidents_persona_dataset.csv')

# Standardise strings
df['effort'] = df['effort'].str.strip().str.lower()
df['psych_safety'] = df['psych_safety'].str.strip().str.lower()
df['action_quality'] = df['action_quality'].str.strip().str.lower()

# Label encoding
effort_map = {'low': 0, 'medium': 1, 'high': 2}
safety_map = {'low': 0, 'mixed': 1, 'medium': 2, 'high': 3}
quality_map = {'poor': 0, 'good': 1}

df['effort_encoded'] = df['effort'].map(effort_map)
df['psych_safety_encoded'] = df['psych_safety'].map(safety_map)
df['action_quality_encoded'] = df['action_quality'].map(quality_map)

# Filter rows where all labels are defined
df = df.dropna(subset=['effort_encoded', 'psych_safety_encoded', 'action_quality_encoded']).copy()
print('✔️ Cleaned dataset shape:', df.shape)
```
